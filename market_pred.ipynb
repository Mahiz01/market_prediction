```python
model=RandomForestClassifier(n_estimators=100,min_samples_split=100,random_state=1)
#defining an instance that is model and n_estimator defines 100 tres should be in a forest and min_samoles_split defines the no of divisions to node and random state defines the values predicted shld be in the range of the 1
```

```python
train=sp500[:-100]
```

```python
predictors=["Close","Open","High","Low","Volume"]
#the parameters that should be checked out
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)   
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
predictions["Target"].value_counts()/predictions.shape[0]
```

```python
horizons=[2,5,60,250,1000]
new_predictors=[]
for horizon in horizons:
    rolling_avg=sp500.rolling(horizon).mean()
    ratio_column=f"close_ratio{horizon}"
    sp500[ratio_column]=sp500["Close"]/rolling_avg["Close"]
    trend_column=f"Trend_{horizon}"
    sp500[trend_column]=sp500.shift(1).rolling(horizon).sum()["Target"]
    new_predictors+=[ratio_column,trend_column]
#rolling avg define the particular range avg of the data set instead of whole dataset
#the horizon is definite point  to which the operstions should be carried out
```

```python
sp500["Close"]
```

```python
sp500=sp500.dropna(subset=sp500.columns[sp500.columns!="Tomorrow"])
```

```python
model=RandomForestClassifier(n_estimators=200,min_samples_split=50,random_state=1)
```

```python
def predict(test,train,predictors,model):
    model.fit(train[predictors],train["Target"])
    pred=model.predict_proba(test[predictors])[:,1]
    pred[pred>=.6]=1
    pred[pred<.6]=0
    pred=pd.Series(pred,index=test.index,name="Predictions")
    combined=pd.concat([test["Target"],pred],axis=1)
    return combined
```

```python
combined.plot()
```

```python
predictions=backtest(sp500,model,new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```
